```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))  # Ancho y alto del gráfico

marcas_clase = [1.05, 2.31, 3.57, 4.83, 6.09]  # Marcas de clase
marcas_texto = ["0.9 - 2.16", "2.17 - 3.43", "3.44 - 4.7", "4.71 - 5.97", "5.98 - 7.24"]  # Etiquetas de las marcas de clase
frecuencias = [8, 10, 6, 4, 2]  # Frecuencias absolutas

plt.bar(marcas_clase, frecuencias, width=1,
        edgecolor="k",
        color=["#33FFBE", "#333CFF", "#FF3349", "#F6FF33", "#9333FF"])
plt.xticks(marcas_clase, marcas_texto, fontsize=10, rotation=45)
plt.xlabel("Marcas de clase", fontsize=25)
plt.ylabel("Frecuencia absoluta", fontsize=25)
plt.title("Histograma", fontsize=40)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 5))  # Ancho y alto del gráfico

marcas_clase = [1, 2, 3, 4, 5]  # Datos del eje x
frecuencias = [8, 18, 24, 28, 30]  # Datos del eje y
marcas_texto = ["0.9 - 2.16", "2.17 - 3.43", "3.44 - 4.7", "4.71 - 5.97", "5.98 - 7.24"]

# Ajustes para el graficado del polígono
datos_x = [0] + marcas_clase
datos_y = [0] + frecuencias

plt.plot(datos_x, datos_y,
         linewidth=5, color="g", linestyle="--",
         marker="v", markersize=20, markerfacecolor="y", markeredgecolor="r")

plt.xticks(marcas_clase, marcas_texto, fontsize=10, rotation=45)
plt.xlabel("Marcas de clase", fontsize=25)  # Etiqueta del eje x
plt.ylabel("Frecuencia acumulada", fontsize=25)  # Etiqueta del eje y
plt.title("Ojiva", fontsize=40)  # Etiqueta del título
plt.grid()  # Activar cuadrícula
plt.show()  # Mostrar gráfico
```
